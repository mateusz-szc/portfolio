```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
import warnings
warnings.filterwarnings("ignore")
```

```python
%store -r descr
```

```python
file = "BLI_18052021173439395.csv"
df = pd.read_csv(file)
df.head()
```

```python
df = df[df.Inequality == "Total"]
df = df[["Country", "Indicator", "Unit", "Value"]]
df
```

```python
df.shape
```

```python
newdf = df[["Indicator", "Unit"]].drop_duplicates(subset="Indicator").reset_index(drop=True)
newdf.Indicator.unique()
```

```python
lista = []

for i in df.Indicator.unique():
    temp_df = df[df.Indicator == i]
    unit = temp_df.Unit.unique()[0]
    temp_df = temp_df[["Country", "Value"]].sort_values(by="Value", ascending=False)
#     temp_df = pd.concat([temp_df.head(), temp_df.tail()])
    
    
    lista.append([i, unit, temp_df])
```

```python
color_palette = ["Spectral", "Wistia_r", "rocket", "magma_r", "viridis", "viridis_r", "YlOrBr_r", "icefire_r"]


plt.rcParams.update({'font.size': 13})
for x in range(len(lista)):
    plt.figure(figsize=(12, 12))
    splot=sns.barplot(x=lista[x][2].Value, y=lista[x][2].Country, palette=random.choice(color_palette))
    for p in splot.patches:        
        width = p.get_width()
        splot.text(width + 0.2, 
                   p.get_y() + p.get_height() / 2,
                   '{:1.2f}'.format(width),
                   ha = 'left',   
                   va = 'center')

    splot.set(ylabel=None)
    splot.set(xlabel=lista[x][1])
    plt.xticks(rotation=45)
    plt.title(f"\n\n{lista[x][0]}\n\n{descr[x]}\n")
    plt.grid()
```
